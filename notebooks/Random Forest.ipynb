```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Loading dataset
file_path = "../data/raw/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv"
df = pd.read_csv(file_path)

df.shape, df.columns[:15]
df.columns
df.columns = df.columns.str.strip()
df.columns[:15]
df['Label'].value_counts()

# Basic info about dataset
df.shape          # total rows and columns
df.info()         # data types and missing values
df.isnull().sum().sort_values(ascending=False).head(10)  # missing columns

# Important features
df[['Flow Duration', 'Total Fwd Packets', 'Total Length of Fwd Packets']].describe()

# checking label distribution
df['Label'].value_counts()

# Converting labels into binary:
# BENIGN = 0, everything else = 1 (attack)
df['Label_binary'] = df['Label'].apply(lambda x: 0 if x == 'BENIGN' else 1)
df['Label_binary'].value_counts()

# Separate features and target
X = df.drop(columns=['Label', 'Label_binary'])
y = df['Label_binary'].values
X = X.replace([np.inf, -np.inf], np.nan)
# Check shapes
X.shape, y.shape

#Spliting for training
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

X_train.shape, X_test.shape

# Making copies so original data stays unchanged
X_train_clip = X_train.copy()
X_test_clip  = X_test.copy()

#Clipping large values
X_train_clip = X_train_clip.clip(lower=-1e12, upper=1e12)
X_test_clip  = X_test_clip.clip(lower=-1e12, upper=1e12)
X_train_clip.describe().T[['min', 'max']].head()

# Handling missing values using median 
imputer = SimpleImputer(strategy='median')
X_train_imputed = imputer.fit_transform(X_train_clip)
X_test_imputed  = imputer.transform(X_test_clip)


# Random Forest Model


from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Using imputed data (no need to scale for Random Forest)
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=20,
    random_state=42,
    n_jobs=-1,               
    class_weight='balanced' 
)



rf.fit(X_train_imputed, y_train)
y_pred = rf.predict(X_test_imputed)

# Print evaluation metrics
print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall   :", recall_score(y_test, y_pred))
print("F1-score :", f1_score(y_test, y_pred))
print()

# Detailed classification report
print(classification_report(y_test, y_pred))

# Feature importance 
feature_names = X.columns
importances = pd.Series(rf.feature_importances_, index=feature_names).sort_values(ascending=False)

# Top 15 important features
importances.head(15)
```
